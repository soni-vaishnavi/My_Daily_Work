```python
class Read:
    def __init__(self, file_name):
        with open (file_name,  "r") as file:
            self.data = file.readlines()
    def ip(self):
        self.ipa = list(map(lambda x : x.split(" ")[0], self.data))
        return self.ipa
    def selected_add(self):
        self.add_less50 = list(filter(lambda x : int(x.split(".")[0])<50, self.ipa))
        return self.add_less50
```

```python
r1 = Read("example_log.txt")
```

```python
r1.ip()
```
